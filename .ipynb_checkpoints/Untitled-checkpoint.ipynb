```python
import numpy as np
import pygame
import sys
import math
WHITE= (255,255,255)
CRIMSON = (220,20,60)
GREEN = (173,255,47)
BLUE = (0,206,209)
BLACK = (0,0,0)
ROW_COUNT = 6
COLUMN_COUNT = 7
count = 1
tie_gameover=False 

def creatingmatrix():
    return np.zeros((ROW_COUNT,COLUMN_COUNT))

def dropball(board,row,col,ball):
    global count
    board[row][col]= ball
    count += 1
    count_method(count)
 
def truelocation(board,col):
    return board[ROW_COUNT-1][col]== 0
 
def get_next_open_row(board,col):
    for r in range(ROW_COUNT):
        if board[r][col]== 0:
            return r

def count_method(total_turn):
    if (total_turn>42):
        label = myfont.render("TIE! PLAY AGAIN", 1,BLACK)
        screen.blit(label, (40,10))
        global tie_gameover
        tie_gameover=True

def printingboard(board):
    return np.flip(board,0)

def winningConditions(board, ball):
    for c in range(COLUMN_COUNT-3):
        for r in range(ROW_COUNT):
            if board[r][c] == ball and board[r][c+1] == ball and board[r][c+2] == ball and board[r][c+3] == ball:
                return True 
    for c in range(COLUMN_COUNT):
        for r in range(ROW_COUNT-3):
            if board[r][c] == ball and board[r+1][c] == ball and board[r+2][c] == ball and board[r+3][c] == ball:
                return True
    for c in range(COLUMN_COUNT-3):
        for r in range(ROW_COUNT-3):
            if board[r][c] == ball and board[r+1][c+1] == ball and board[r+2][c+2] == ball and board[r+3][c+3] == ball:
                return True
    for c in range(COLUMN_COUNT-3):
        for r in range(3, ROW_COUNT):
            if board[r][c] == ball and board[r-1][c+1] == ball and board[r-2][c+2] == ball and board[r-3][c+3] == ball:
                return True 

def draw_board(board):
    for c in range(COLUMN_COUNT):
        for r in range(ROW_COUNT):
            pygame.draw.rect(screen, WHITE, (c*SQUARESIZE, r*SQUARESIZE+SQUARESIZE, SQUARESIZE, SQUARESIZE))
            pygame.draw.circle(screen, CRIMSON, (int(c*SQUARESIZE+SQUARESIZE/2), int(r*SQUARESIZE+SQUARESIZE+SQUARESIZE/2)), RADIUS)
    
    for c in range(COLUMN_COUNT):
        for r in range(ROW_COUNT):      
            if board[r][c] == 1:
                pygame.draw.circle(screen, GREEN, (int(c*SQUARESIZE+SQUARESIZE/2), height-int(r*SQUARESIZE+SQUARESIZE/2)), RADIUS)
            elif board[r][c] == 2: 
                pygame.draw.circle(screen, BLUE, (int(c*SQUARESIZE+SQUARESIZE/2), height-int(r*SQUARESIZE+SQUARESIZE/2)), RADIUS)
    pygame.display.update()
        
board = creatingmatrix()
print(printingboard(board))
game_over = False
turn = 0

pygame.init()
SQUARESIZE = 75
width = COLUMN_COUNT * SQUARESIZE
height = (ROW_COUNT+1) * SQUARESIZE
size = (width, height)
RADIUS = int(SQUARESIZE/2 - 5)
screen = pygame.display.set_mode(size)
draw_board(board)
pygame.display.update()

myfont = pygame.font.SysFont("Baskerville Old Face", 55)     

while not game_over:   
   if tie_gameover:
        pygame.time.wait(3000)
        sys.exit()  
   for event in pygame.event.get():
        if event.type == pygame.QUIT:
            sys.exit()

        if event.type == pygame.MOUSEMOTION:
            pygame.draw.rect(screen, CRIMSON, (0,0, width, SQUARESIZE))
            posx = event.pos[0]

            if turn == 0:
                pygame.draw.circle(screen, GREEN, (posx, int(SQUARESIZE/2)), RADIUS)
            else: 
                pygame.draw.circle(screen,BLUE, (posx, int(SQUARESIZE/2)), RADIUS)
        pygame.display.update()

        if event.type == pygame.MOUSEBUTTONDOWN:
            pygame.draw.rect(screen, CRIMSON, (0,0, width, SQUARESIZE))

            if turn == 0:
                posx = event.pos[0]
                col = int(math.floor(posx/SQUARESIZE))

                if truelocation(board, col):
                    row = get_next_open_row(board, col)
                    dropball(board, row, col, 1)

                    if winningConditions(board, 1):
                        label = myfont.render("Player 1 wins!!", 1, GREEN)
                        screen.blit(label, (105,10))
                        game_over = True

            elif turn==1:               
                posx = event.pos[0]
                col = int(math.floor(posx/SQUARESIZE))   

                if truelocation(board, col):
                    row = get_next_open_row(board, col)
                    dropball(board, row, col, 2)
        
                    if winningConditions(board, 2):
                        label = myfont.render("Player 2 wins!!", 1, BLUE)
                        screen.blit(label, (105,10))
                        game_over = True 

            printingboard(board)
            draw_board(board)

            turn += 1
            turn = turn % 2

            if game_over:
                pygame.time.wait(7000)




   

  

 

   

```
